```python
import pandas as pd
import numpy as np

np.random.seed(42)

dates = pd.date_range(start="2023-01-01", periods=100)

data = {
    "Date": dates,
    "Open": np.random.uniform(2400, 2600, size=100).round(2),
}

data["High"] = (data["Open"] + np.random.uniform(10, 50, size=100)).round(2)
data["Low"] = (data["Open"] - np.random.uniform(10, 50, size=100)).round(2)
data["Close"] = (data["Low"] + np.random.uniform(5, 50, size=100)).round(2)
data["Volume"] = np.random.randint(1000000, 5000000, size=100)

df = pd.DataFrame(data)

df.to_csv("reliance_synthetic_data.csv", index=False)

df.head()
```

```python

import pandas as pd

df = pd.read_csv("reliance_synthetic_data.csv")

print("Dataset shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nData Types:\n", df.dtypes)
print("\nMissing Values:\n", df.isnull().sum())
df.head()
```

```python

df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

plt.figure(figsize=(8, 5))
sns.histplot(df['Close'], bins=20, kde=True)
plt.title("Distribution of Close Prices")
plt.xlabel("Close Price")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Volume'])
plt.title("Box Plot of Trading Volume")
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(x='Open', y='Close', data=df)
plt.title("Open vs Close Price")
plt.xlabel("Open")
plt.ylabel("Close")
plt.show()
```

```python

df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Close', data=df)
plt.title("Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.xticks(rotation=45)
plt.show()
```
